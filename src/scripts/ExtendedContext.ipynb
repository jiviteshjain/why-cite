```python
extended_contexts = []
working_rows = []

for index,row in df.iterrows():
    
    print("\n******** ",index,"\n")
        
#     if index%100 == 0: print(index)
        
    core_id = str(row["core_id"])
    f = open(f"./train/{core_id}.txt")
    data = f.read()
    sent = tokenize.sent_tokenize(data)
    cit = row["citation_context"]
    print("Citation: ",cit)
    
#     cit = "The State of the Carbon Cycle Report"
    
    cit = re.sub('[^A-Za-z ]+', ' ', cit)
    r_cit = cit.replace('AUTHOR TAG','(.*?)')
    cit = cit.replace('AUTHOR TAG',' ')
    cit = re.sub(' +', '', cit)
    r_cit = re.sub(' +', '', r_cit)
    
    print("\nPreprocessed citation: ",index," ",cit)
    print("\nPreprocessed regex citation: ",index," ",r_cit,"\n")
    
    
    ind = 0
    for i in range(len(sent)):
                
        s_p = ""
        s_a = ""
        if i != 0: s_p = sent[i-1]
        s = sent[i]
        if i != (len(sent)-1): s_a = sent[i+1]
        s = s_p + s + s_a
        
        temps = s
        s = re.sub(r'http\S+', '', s)
        s = re.sub('[^A-Za-z ]+', ' ', s)
        s = s.replace("\n","")
        s = re.sub(' +', '', s)
        
        if cit in s: 
            print("Matched with cit in s: ")
            print(s)
            break
            
        if s in cit and len(s) != 1:
            print("Matched with s in cit: ")
            print(temps)
            print(len(s))
            break
        
        match = re.search(r_cit,s)
        
        if match is not None:
            print("Matched with regex: ")
            print(match.group(0))
            break
        
        
        ind += 1
    
#     print(ind)

    
    try:
        extended_context = ""
        extended_context += (sent[ind].replace("\n",""))
        extended_context += (sent[ind+1].replace("\n",""))
        extended_context += (sent[ind+2].replace("\n",""))
        print("\nExtended Context: ",extended_context)
#         print("\n",sent[ind+1])
#         print("\n",sent[ind+2])
        
        extended_contexts.append(extended_context)
        working_rows.append(index)
        
    except:
        print("didnt work",index, " ", cit, " ", core_id)
        extended_contexts.append(row["citation_context"])
        continue
        
#     if index == 11: break
```
